```python
import pandas as pd

import sqlite3
from sqlite3 import Error
```

```python
SWISS = pd.read_csv("USCHF.csv")
JAP = pd.read_csv("USJP.csv")
EUR = pd.read_csv("USEU.csv")
BRIT = pd.read_csv("USGB.csv")
```

```python
SWISS.columns =['month','av-USD-CHF','min-USD-CHF','max-USD-CHF','nb-days']
JAP.columns =['month','av-USD-JPY','min-USD-JPY','max-USD-JPY','nb-days']
EUR.columns =['month','av-USD-EUR','min-USD-EUR','max-USD-EUR','nb-days']
BRIT.columns =['month','av-USD-GBP','min-USD-GBP','max-USD-GBP','nb-days']
```

```python
SWISS['month']=SWISS['month'].str.replace('/','-')
new = SWISS['month'].str.split("-", n = 1, expand = True)
SWISS["Month"]= new[0] 
SWISS["Year"]= new[1] 
SWISS.drop(columns =["month"], inplace = True) 

SWISS['date'] = SWISS["Year"].map(str) + "-" + SWISS["Month"].map(str)
SWISS.drop(columns =["Month","Year"], inplace = True) 
SWISS
```

```python
EUR['month']=EUR['month'].str.replace('/','-')
new = EUR['month'].str.split("-", n = 1, expand = True)
EUR["Month"]= new[0] 
EUR["Year"]= new[1] 
EUR.drop(columns =["month"], inplace = True) 

EUR['date'] = EUR["Year"].map(str) + "-" + EUR["Month"].map(str)
EUR.drop(columns =["Month","Year"], inplace = True) 
EUR
```

```python
BRIT['month']=BRIT['month'].str.replace('/','-')
new = BRIT['month'].str.split("-", n = 1, expand = True)
BRIT["Month"]= new[0]
BRIT["Year"]= new[1] 
BRIT.drop(columns =["month"], inplace = True) 

BRIT['date'] = BRIT["Year"].map(str) + "-" + BRIT["Month"].map(str)
BRIT.drop(columns =["Month","Year"], inplace = True) 
BRIT
```

```python
def create_connection(db_file):
    """ create a database connection to a SQLite database """
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
    finally:
        conn.close()
 
if __name__ == '__main__':
    create_connection("currency.sqlite")
```

```python
conn = sqlite3.connect("currency.sqlite")
cur = conn.cursor()
EUR.to_sql(name = "EUR", con = conn)
query = "SELECT * FROM EUR;"
df = pd.read_sql_query(query,conn)
df
```
