```python
import pandas as pd

import sqlite3
from sqlite3 import Error
 
```

```python
Currency_exchange = pd.read_csv("new.csv")
Currency_exchange.columns =['month','av-USD-EUR','min-USD-EUR','max-USD-EUR','nb-days']
Currency_exchange['month']=Currency_exchange['month'].str.replace('/','-')
new = Currency_exchange['month'].str.split("-", n = 1, expand = True)
Currency_exchange["Month"]= new[0] 
  
# making seperate last name column from new data frame 
Currency_exchange["Year"]= new[1] 
  
# Dropping old Name columns 
Currency_exchange.drop(columns =["month"], inplace = True) 

Currency_exchange['date'] = Currency_exchange["Year"].map(str) + "-" + Currency_exchange["Month"].map(str)
Currency_exchange['date'] = pd.to_datetime(Currency_exchange['date'],format="%Y/%m")
Currency_exchange.drop(columns =["Month","Year"], inplace = True) 
Currency_exchange
```

```python
def create_connection(db_file):
    """ create a database connection to a SQLite database """
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
    finally:
        conn.close()
 
if __name__ == '__main__':
    create_connection("new.sqlite")
```

```python
conn = sqlite3.connect("new.sqlite")
cur = conn.cursor()
Currency_exchange.to_sql(name = "currency5", con = conn)
```
