```python
import requests
page = requests.get("http://fxtop.com/en/historical-exchange-rates.php?A=1&C1=USD&C2=CHF&DD1=01&MM1=01&YYYY1=1953&B=1&P=&I=1&DD2=30&MM2=11&YYYY2=2018&btnOK=Go%21")
page.content
import pandas as pd
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify)
```

```python
pd.read_html("http://fxtop.com/en/historical-exchange-rates.php?A=1&C1=USD&C2=EUR&DD1=01&MM1=01&YYYY1=1953&B=1&P=&I=1&DD2=30&MM2=11&YYYY2=2018&btnOK=Go%21")
```

```python
table = soup.find_all('table', {'border':1})
```

```python
table
```

```python
USA_CHF_data = pd.read_html(str(table), index_col=0, header = 0)
```
