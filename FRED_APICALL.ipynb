```python
data = fred.observations('USREC')
```

```python
import pandas as pd
recession_index_df = pd.DataFrame.from_dict(data)
```

```python
recession_index_df.head(500)
```

```python
ob_recession_df = recession_index_df['observations']
ob = pd.DataFrame.from_dict(ob_recession_df, orient='columns')
ob.head(1)
```

```python
dates = []
values = []

for x in range(len(ob)):
    a = ob['observations'][x]['date']
    b = ob['observations'][x]['value']
    dates.append(a)
    values.append(b)
```

```python
observation_rec = pd.DataFrame(
    {'Date': dates,
     'Recession_Indicator': values
     })
```

```python
observation_rec.head()
```

```python
observation_rec.to_csv('recession.csv')
```
